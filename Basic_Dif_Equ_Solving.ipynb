```python
import numpy as np
from scipy.integrate import odeint
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def firstorder(y,t,K,u):
    tau =5.0
    K = 2.0
    u = 1.0
    dydt =(-y +K*u)/tau
    return dydt
    
```

```python
t= np.linspace(0,10,11)
K = 2.0
u = 1.0
```

```python
y = odeint(firstorder,0,t,args=(K,u))
```

```python
plt.plot(t,y)
plt.title("Solving First Order D.E. in Python")
plt.xlabel("time")
plt.ylabel("values")
plt.show()
```

```python
import numpy as np
import pandas as pd
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def model(y,t):
    k = 10
    dydt = -t*y / k
    return dydt
y0 = 5


t = np.linspace(0,20,2)

y = odeint(model,y0,t)

plt.plot(t,y)
plt.xlabel('time')
plt.ylabel('y(t)')
plt.show()

Y = np.squeeze(np.asarray(y))
T = np.squeeze(np.asarray(t))
print(Y)
print(T)
```
